```python
# Anlisis de la relacin entre 'Impresiones' y 'Porcentaje de clics de las impresiones'. EDA 1.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('YouTube SirAlatriste datos.csv')

# un scatter plot con seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Impresiones', y='Porcentaje de clics de las impresiones (%)', color='#4869e3')

# lnea de tendencia
sns.regplot(data=df, x='Impresiones', y='Porcentaje de clics de las impresiones (%)', scatter=False)

# ttulos y etiquetas
plt.title('Relacin entre Impresiones y Porcentaje de clics de las impresiones (%)')
plt.xlabel('Impresiones')
plt.ylabel('Porcentaje de clics de las impresiones (%)')
plt.ylim(0, 15)
plt.xlim(0, 75000)

plt.show()
```

```python
#EDA 2. Todos los videos del canal, fecha.
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

df = pd.read_csv('YouTube SirAlatriste datos.csv')

fechas = df['Fecha']

#convierte las fechas a objetos de fecha y hora
fechas = [datetime.strptime(fecha, '%b %d, %Y') for fecha in fechas]

#extrae el ao de cada fecha
aos = [fecha.year for fecha in fechas]

#aqui se crea una serie de pandas con los aos
aos = pd.Series(aos)

media = aos.value_counts().mean()

#para crear un grfico de barras con los aos
aos.value_counts().sort_index().plot(kind='bar', x='Videos realizados', y= 'Ao')

#Dibujar una lnea horizontal en el valor de la media
plt.axhline(media, color='#11CA38', linestyle='--')

# as se agrega una etiqueta con el valor de la media
plt.text(5, media + 0.5, f'Media: {media:.0f}', color='#11CA38')
plt.text(5.6, 4.5, 'Irlanda', color='black')
plt.text(4.75, 27.5, 'TFG', color='black')

plt.ylabel('Videos realizados por ao')

plt.show()
```

```python
#Anlisis de la duracin media de las visualizaciones. EDA 6.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv('YouTube SirAlatriste datos.csv')

# para ordenar los valores del eje x
datos_sorted = datos.sort_values('Duracin media de las visualizaciones')

plt.figure(figsize=(10, 10))

# histograma con Seaborn

sns.histplot(data=datos_sorted, x='Duracin media de las visualizaciones', binwidth=0.5, binrange=(0, 18), color='#4869e3')

# para agregar un ttulo al grfico
plt.title('Anlisis de la duracin media de las visualizaciones')

media = datos['Duracin media de las visualizaciones'].mean()
mediana = datos['Duracin media de las visualizaciones'].median()

plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana, color='#11CA38', linestyle='dashed', linewidth=2, label='Mediana')

# agregar etiquetas a los ejes
plt.xlabel('Duracin media de las visualizaciones (en minutos)')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0);
plt.xlim(0, 17.5)
plt.ylim(0, 62.5)
plt.legend()

plt.show()
```

```python
#EDA 15. Playlist del canal. Cuales han sido mejores en funcin de varios datos
import plotly.express as px
import pandas as pd

df = pd.read_csv('YouTube SirAlatriste Playlists.csv')

# lista con las variables de color, osea con aquellas columnas que mandan los datos
color_vars = ['Tiempo de visualizacin de la lista de reproduccin (horas)', 'Duracin media de las visualizaciones de la lista de reproduccin', 'Visualizaciones de la lista de reproduccin']

# lista con los treemaps para cada variable de color
treemaps = []
for color_var in color_vars:
    # esto cambia el DataFrame en funcin de la variable de color de cada variable.. he conseguido ponerlo en orden y cuadrar el texto en el centro de la celda
    df_sorted = df.sort_values(by=color_var)
    
    treemap = go.Treemap(
        labels=df_sorted['Ttulo de la lista de reproduccin'],
        parents=[''] * len(df),
        values=df_sorted[color_var],
        marker=dict(colors=df_sorted[color_var], colorscale='Blues'),
        textinfo='label+value',
        textposition='middle center',
        name=color_var)
    
    treemaps.append(treemap)
    

# MEN DESPEGABLE

updatemenus = [dict(
    type='dropdown',
    showactive=True,
    buttons=[dict(
        label=color_var,
        method='update',
        args=[{'visible': [color_var == c for c in color_vars]},
              {'data': [treemaps[color_vars.index(color_var)]]}]
    ) for color_var in color_vars])]

# layout del grfico
layout = go.Layout(updatemenus=updatemenus)

fig = go.Figure(data=treemaps, layout=layout)

fig.show()

plotly.offline.plot(fig, filename='grafico 3 plotly.html')
```
